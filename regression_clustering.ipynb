```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {'area': [35, 40, 47, 53, 58, 62, 68, 75, 83, 89, 95, 101, 112, 122, 130],
        'price': [40000, 43000, 50000, 53000, 61000, 70000, 75000, 82000, 88000, 95000, 100000, 115000, 125000, 138000, 150000]}
df = pd.DataFrame(data)

model = LinearRegression()
X_housing = df[['area']]
y_housing = df['price']
model.fit(X_housing, y_housing)

prediction = model.predict(pd.DataFrame({'area': [90]}))
print(f'Прогноз цены для 90 кв.м.: {prediction[0]:.2f}')

X_line = pd.DataFrame({'area': range(30, 140)})
y_line = model.predict(X_line)

plt.scatter(X_housing, y_housing)
plt.plot(X_line, y_line)
plt.xlabel('площадь')
plt.ylabel('цена')
plt.title('прогнозирование цен')
plt.show()
```

## Вывод по заданию 1

Модель предсказала цену ~99 585 для квартиры 90 кв.м.
Это логично, так как соседние значения в датасете (83 кв.м. → 88 000, 89 кв.м. → 95 000) подтверждают этот диапазон.
Чем больше площадь — тем выше цена, что видно на графике.

```python
from sklearn.cluster import KMeans

# Данные: [Возраст, Сумма трат в месяц]
customers = [[18, 50], [20, 60], [45, 10], [50, 15], [25, 55], [55, 5], [22, 45], [60, 12], [24,57], [35,33], [42,8], [65,8], [30,39]]

kmeans = KMeans(n_clusters = 3)

labels = kmeans.fit_predict(customers)
print(labels)

X_customers = [p[0] for p in customers]
y_customers = [p[1] for p in customers]
plt.scatter(X_customers, y_customers, c = labels)
plt.xlabel('возраст')
plt.ylabel('траты')
plt.title('кластеризация клиентов')
plt.show()
```

## Вывод по заданию 2

Модель разделила клиентов на три кластера:
1. Молодые (приблизительно 20-25 лет) которые много тратят (45+)
2. Взрослые (40+) которые мало тратят (до 20)
3. Среднее / что-то между. Люди возраста около 30 которые тратят около 35

В целом сегментация логичная и график это подтверждает.
